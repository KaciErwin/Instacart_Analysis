## Finding department orders

```python
# find average order number for each department
df_merged.groupby('department_id')['order_number'].mean()
```

## Finding customer loyalty flags

```python
# create table with maximum customer orders column
df_merged['max_order']=df_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check table shape
df_merged.shape
```

```python
# check table 
df_merged.head()
```

```python
# check counts for maximum customer orders
df_merged['max_order'].value_counts()
```

```python
# Create new column with customer loyalty flag
df_merged.loc[df_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
df_merged.loc[(df_merged['max_order'] <= 40) & (df_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
df_merged.loc[df_merged['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# check table
df_merged.head()
```

```python
# check loyalty flag counts
df_merged['loyalty_flag'].value_counts(dropna = False)
```

## Check loyalty customer statistics

```python
# Create a list of price statistics based on customer loyalty
df_merged.groupby('loyalty_flag').agg({'prices' : ['mean', 'max', 'min']})
```

## Create a spending flag for customers

```python
# create a column for average spent by user
df_merged['spending_mean']=df_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check table
df_merged.head()
```

```python
# create a spending flag column 
df_merged.loc[df_merged['spending_mean'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
df_merged.loc[df_merged['spending_mean'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
# check table
df_merged.head()
```

```python
# check spending flag counts
df_merged['spending_flag'].value_counts(dropna = False)
```

## Create a frequency flag for customers

```python
# create frequency median column
df_merged['frequency_median'] = df_merged.groupby('user_id')['days_since_prior_order'].transform('median')
```

```python
# create the frequency flag column
df_merged.loc[df_merged['frequency_median'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
df_merged.loc[(df_merged['frequency_median'] <= 20) & (df_merged['frequency_median'] >10), 'frequency_flag'] = 'Regular Customer'
```

```python
df_merged.loc[df_merged['frequency_median'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# check table
df_merged.head()
```

```python
# check frequency flag value counts
df_merged['frequency_flag'].value_counts(dropna = False )
```
